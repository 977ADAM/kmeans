```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons, make_circles, make_blobs
from sklearn.cluster import KMeans, DBSCAN

plt.figure(figsize=(5, 5))
```

```python
# X, y = make_moons(n_samples=100, random_state=10)
# X, y = make_circles(n_samples=100, factor=0.5, noise=0.05, random_state=10)
X, y = make_blobs(n_samples=100, random_state=10)

df = pd.DataFrame(X, columns=['x1', 'x2'])
df['target'] = y

print(df.head())

plt.scatter(df["x1"], df["x2"])
```

```python
model = KMeans(n_clusters=3, random_state=48, init="k-means++", n_init=10)
```

```python
model.fit(X)
```

```python
l = model.predict(X)
```

```python
plt.figure(figsize=(8, 6))
for i in np.unique(l):
    plt.scatter(X[l == i][:, 0], X[l == i][:, 1], label=f"Cluster {i}")
plt.legend()
```

```python
plt.figure(figsize=(8, 6))
for i in np.unique(l):
    plt.scatter(X[l == i][:, 0], X[l == i][:, 1], label=f"Cluster {i}")

for i in model.cluster_centers_:
    plt.scatter(i[0], i[1], marker="*", s=200, c="black", label="Centroid")
plt.legend()
```
